```python
def setPltStyle(cmap='viridis', linestyle='-', font='serif', fontsize=20, linecolors=['blue', 'red', 'green', 'purple', 'orange']):
    plt.set_cmap(cmap)
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=linecolors)
    plt.rcParams['lines.linestyle'] = linestyle
    plt.rcParams['font.family'] = font
    plt.rcParams['font.size'] = fontsize

def setTitles(title=None, xtitle=None, ytitle=None):
    if title is not None:
        plt.title(title)
    if xtitle is not None:
        plt.xlabel(xtitle)
    if ytitle is not None:
        plt.ylabel(ytitle)
        
def setColors():
    colormap = plt.cm.get_cmap('plasma')
    colors = colormap([0,100])
    
```
