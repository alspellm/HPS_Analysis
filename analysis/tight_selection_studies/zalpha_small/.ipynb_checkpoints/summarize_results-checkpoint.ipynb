```python
#Graph flat z0 cut values
masses = []
cuts = []
zbis = []
nsigs = []
nbkgs = []
effs = []
for i,file_path in enumerate(root_files):
    #if i > 2:
    #    break
    mass = int(float(os.path.basename(file_path).split('_')[1]))
    #if mass != 55:
    #    continue
    print(mass)
    
    with uproot.open(file_path) as file:
        slope = 0.0
        cut_values = file[f'slope_{slope}/cut_values_g'].values()
        zbi_values = file[f'slope_{slope}/zbis_g'].values()
        nbkg_values = file[f'slope_{slope}/nbkg_g'].values()
        nsig_values = file[f'slope_{slope}/nsig_g'].values()
        eff_values = file[f'slope_{slope}/eff_g'].values()
        peak_idx = np.where(zbi_values[1] == np.max(zbi_values[1]))[0][0]
        #print('PEAK INDEX', peak_idx)
        
        cut = cut_values[1][peak_idx]
        #print('Cut Value:', cut)
        nbkg = nbkg_values[1][peak_idx]
        nsig = nsig_values[1][peak_idx]
        zbi = zbi_values[1][peak_idx]
        eff = eff_values[1][peak_idx]
        
        cuts.append(cut)
        zbis.append(zbi)
        nsigs.append(nsig)
        nbkgs.append(nbkg)
        effs.append(eff)
        masses.append(mass)

fig, ax = plt.subplots(figsize=(15,10))
#fig, ax = plt.subplots()
ax.plot(masses, nsigs, label='Nsig(x5)',color='blue')
ax.plot(masses, nbkgs, label='Nbkg(10% Data)',color='red')
#ax.set_yscale('log')
ax.set_ylabel('Nsig(x5) and Nbkg', fontsize = 20)
ax.set_xlabel("VD Mass [MeV]", fontsize = 20)
ax.tick_params(axis='y', labelsize = 20)
ax.tick_params(axis='x', labelsize = 20)
ax.set_ylim(bottom=0.0, top=50)
ax.yaxis.grid(True, linestyle='--', which='major', color='gray', alpha=0.7)

ax2 = ax.twinx()
ax2.plot(masses, zbis, label='ZBi',color='green')
ax2.set_ylabel('ZBi', color='green', fontsize = 20)
ax2.tick_params(axis='y',labelcolor='green', labelsize = 20)
ax2.set_ylim(bottom=0.0, top=6)
#ax2.text(1,6,f'zalpha_slope_{zalpha_slope}')

fig.legend(loc='center', fontsize=20)
plt.title('Flat Z0 Cut Max ZBi', fontsize=20)
fig.savefig(f'{outdir}/flat_z0_cut_results.png')

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(masses, cuts, label='best zbi')
ax.set_ylabel('Z0 [mm]', fontsize = 20)
ax.set_xlabel("VD Mass [MeV]", fontsize = 20)
ax.tick_params(axis='y', labelsize = 20)
ax.tick_params(axis='x', labelsize = 20)
ax.yaxis.grid(True, linestyle='--', which='major', color='gray', alpha=0.7)
plt.title('Flat Z0 Cut Values', fontsize=20)
fig.savefig(f'{outdir}/flat_z0_cut_values.png')
```

```python
from scipy.optimize import curve_fit

# Define a linear function (slope * x + intercept)
def linear_func(x, slope, intercept):
    return slope * x + intercept

masses = np.array(masses, dtype=float)[masses <= 120.]
cuts = cuts[:len(masses)]
print(masses)
# Fit the data using curve_fit
params, covariance = curve_fit(linear_func, masses, cuts)

# Get the fitted parameters
fit_slope, fit_intercept = params

# Plot the data
fig, ax = plt.subplots()
ax.scatter(masses, cuts, label='Cut Values')
ax.plot(masses, linear_func(masses, fit_slope, fit_intercept), color='red', label='Fitted Line')
plt.title('Flat Z0 Cut Values')
ax.set_xlabel('VD Mass [GeV]')
ax.set_ylabel('Z0 Cut Value [mm]')
ax.text(40,0.4,f'Slope:{fit_slope} \n Yint:{fit_intercept}')
fig.savefig(f'{outdir}/flat_z0_cut_values_linear_fit.png')
```
